```python
import sympy as sm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
x_primes = [i for i in sm.sieve.primerange(1, 500000)]
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(x_primes, bins=50)
plt.xticks(np.arange(0, 500001, 10000), rotation='vertical');
```

```python
p=[]
upper_limit=1000

for x in range(2,upper_limit):
    if sm.ntheory.primetest.isprime(x):
        p.append(x)

t=np.linspace(2,upper_limit,1000)
y1=1/np.log(t)
y2=[sm.ntheory.generate.primepi(t)/t for t in range(upper_limit)]

fig,axis=plt.subplots(1,2,figsize=(15,5))

plt.subplot(1,2,1)
axis[0].set_title('Asymptotic nature of the prime density function\n is approximately the logarithmic integral')
axis[0].plot(t,y1)
axis[0].plot(t,y2)
axis[0].legend(['li(x)','$\dfrac{\pi(x)}{x}$'],loc='best')

y3=[sm.ntheory.generate.primepi(t) for t in range(upper_limit)]
y4=t/np.log(t)

plt.subplot(1,2,2)
axis[1].set_title('Asymptotic nature of the prime counting function\n is approximately the inverse of the logarithmic integral')
axis[1].plot(t,y3)
axis[1].plot(t,y4)
axis[1].legend(['$\pi(x)$','$\dfrac{x}{\ln(x)}$'],loc='best')

plt.setp(axis[0],xlabel='Search Size $x$')
plt.setp(axis[0],ylabel='Prime Density')

plt.setp(axis[1],xlabel='Search Size $x$')
plt.setp(axis[1],ylabel='Number of Primes $<x$')

plt.show()

# fig.savefig('Fig1.png',dpi=190)
```
