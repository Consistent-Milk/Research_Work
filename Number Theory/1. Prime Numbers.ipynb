```python
import sympy as sm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
x_primes = [i for i in sm.sieve.primerange(1, 500000)]
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(x_primes, bins=50)
plt.xticks(np.arange(0, 500001, 10000), rotation='vertical');
```
